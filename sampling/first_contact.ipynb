```python
#data example

import pandas as pd
import numpy as np

np.random.seed(42)

n_students = 10000
n_schools = 100

# 학교 ID 할당 (100개 학교)
school_ids = [f"S{str(i).zfill(3)}" for i in range(1, n_schools + 1)]
students_per_school = np.random.multinomial(n_students, [1/n_schools]*n_schools)

data = []

student_id = 1
for school_id, n in zip(school_ids, students_per_school):
    for _ in range(n):
        grade = np.random.choice([1, 2, 3])
        gender = np.random.choice(["M", "F"])
        score = np.random.normal(loc=70, scale=10)
        data.append({
            "student_id": student_id,
            "school_id": school_id,
            "grade": grade,
            "gender": gender,
            "score": round(score, 2)
        })
        student_id += 1

df = pd.DataFrame(data)
print(df.head())
```
